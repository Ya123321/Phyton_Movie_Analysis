```python
type(df["Rating"])
```

```python
df["new_Rating"]=df["Rating"].replace(to_replace='no-rating',value='0')
```

```python
top_10_popular_movies=df[["movie title","new_Rating","User Rating","Director"]].sort_values(by=["new_Rating"], ascending=False)[:10]
top_10_popular_movies
```

```python
plt.figure(figsize=(25,5))
sns.barplot(data=top_10_popular_movies,x="movie title",y="new_Rating")
plt.title("top_10_popular_movies")
plt.show()
```

```python
df["User Rating"]=df["User Rating"].str.replace("K","000")
df["User Rating"]=df["User Rating"].str.replace("M","000000")
```

```python
df["User Rating"]=pd.to_numeric(df["User Rating"])
```

```python
top_5_popular_movies_by_User_Rating=df[["movie title","User Rating"]].sort_values(by="User Rating", ascending=False)[:5]
top_5_popular_movies_by_User_Rating
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="User Rating",y="movie title",data=top_5_popular_movies_by_User_Rating)
plt.title("top_5_popular_movies_by_User_Rating")
plt.show()
```
