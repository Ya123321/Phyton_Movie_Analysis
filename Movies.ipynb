```python
df= pd.read_csv('25k IMDb movie Dataset.csv')
df
```

dropping columns path and Run Time

```python
df=df.drop("path",axis=1)
```

```python
df=df.drop("Run Time",axis=1)
```

separate the names of the casts

```python
Casts= pd.Series(data= df['Top 5 Casts'], dtype="string").str.split(pat=",",expand=True)
Casts
```

```python
Casts=Casts.squeeze(axis=1)
Casts
```

DataFrame.size - Number of elements in the array <br>
Series.dropna - Return a new Series with missing values removed <br>
Series.to_numpy- convert Series to NumPy array <br>
Series.squeeze() - convert NumPy array to Series<br>
numpy.reshape- a, newshape, order='C') <br>
"a" - the numpy included the names of casts <br>
"newshape" will be the number of elements in the array<br>
value_counts()- function is used to get a Series containing counts of unique values.<br>
The resulting object will be in descending order so that the first element is the most frequently-occurring element<br>
Return a Series containing counts of unique rows in the DataFrame


```python
type(Casts)
```

```python
Casts = pd.Series(data=np.reshape(Casts.to_numpy(),Casts.size)).dropna()
Casts.count()
Casts
```

```python
np.reshape(Casts.to_numpy(),Casts.size)
```

```python
Casts_1 =pd.Series(data=np.reshape(Casts.to_numpy(),Casts.size)).dropna().value_counts()
Casts_1
```

Series.to_frame - Convert Series to DataFrame

```python
#Casts_1.to_frame(name="count")
```

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.
DataFrame(data,index,columns) the value_counts become to be the columns "count"

```python
Casts_1 = pd.DataFrame(data=Casts_1,columns=["count"])
Casts_1
```

```python
Casts_1=Casts_1[3:]
Casts_1
```

```python
top_5_engaged_actors=Casts_1.head(5)
top_5_engaged_actors
```

```python
top_5_engaged_actors.plot.barh()
plt.show
```

```python
type(df["Rating"])
```

replace 'no-rating' to be 0 rating

```python
df["new_Rating"]=df["Rating"].replace(to_replace='no-rating',value='0')
```

```python
top_10_popular_movies=df[["movie title","new_Rating","User Rating","Director"]].sort_values(by=["new_Rating"], ascending=False)[:10]
top_10_popular_movies
```

```python
plt.figure(figsize=(25,5))
sns.barplot(data=top_10_popular_movies,x="movie title",y="new_Rating")
plt.title("top_10_popular_movies")
plt.show()
```

```python
df["Generes"].unique()
```

```python
Unwanted_characters=["(",")","I","V"," ","X","-","''","'","[","]"]
for a in Unwanted_characters:
  df["Generes"]=df["Generes"].str.replace(a,"")
```

```python
df["Generes"]
```

```python
Popular_genres= pd.Series(data= df['Generes'], dtype="string").str.split(pat=",",expand=True)
Popular_genres
```

```python
df['Generes'].value_counts()
```

```python
Popular_genres =pd.Series(data=np.reshape(Popular_genres.to_numpy(),Popular_genres.size)).dropna().value_counts()
Popular_genres
```

```python
#Popular_genres.to_frame(name="count")
```

```python
Popular_genres = pd.DataFrame(data=Popular_genres,columns=['count'])
Popular_genres
```

```python
Top_10_frequent_Genre =Popular_genres.head(10)
Top_10_frequent_Genre
```

```python
Top_10_frequent_Genre.plot.barh()
plt.show()
```

```python
df["User Rating"]=df["User Rating"].str.replace("K","000")
df["User Rating"]=df["User Rating"].str.replace("M","000000")
```

```python
df["User Rating"]=pd.to_numeric(df["User Rating"])
```

```python
top_5_popular_movies_by_User_Rating=df[["movie title","User Rating"]].sort_values(by="User Rating", ascending=False)[:5]
top_5_popular_movies_by_User_Rating
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="User Rating",y="movie title",data=top_5_popular_movies_by_User_Rating)
plt.title("top_5_popular_movies_by_User_Rating")
plt.show()
```
